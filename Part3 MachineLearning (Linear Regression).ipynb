```python
# import packages
import numpy as np
import os
import pandas as pd

#Set random seed
np.random.seed(42)

#Fig packages import
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
```

# Import Maching Learning Dataset

```python
ML_Data = pd.read_csv('YourPath/ML_Dataset.csv')
ML_Data = ML_Data.drop(ML_Data.columns[0], axis=1)
ML_Data
```

Extract X and y from ML_Data

```python
X = ML_Data.drop(ML_Data.columns[-1], axis=1)
X
```

```python
y = ML_Data['yield']
y
```

# Data Preprocessing

Use standardscaler to preprocess X.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

# Data Splitting

Randomly split (70/30) data X, y into training set and test set.

To keep the results reproductive, set the randome_state = 42.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# First Model: Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train, y_train)
```

```python
y_pred_lin = lin_reg.predict(X_test)
y_pred_lin
```

Chect the model parameters: RMSE, R2. 

```python
from sklearn.metrics import mean_squared_error
RMSE_lin = np.sqrt(mean_squared_error(y_test, y_pred_lin))
RMSE_lin
```

```python
from sklearn.metrics import r2_score
R2_lin = r2_score(y_test, y_pred_lin)
R2_lin
```

We just finished the training of first linear regression model! RMSE: 15.194004003903915 and R2: 0.7111793296958059.

Now let's make y_pred vs y_test plot.

```python
plt.plot(y_pred_lin, y_test, 'b.')
plt.title('Linear Regression')
plt.axis([-10, 120, -10, 120])
plt.xlabel('Predicted Yield')
plt.ylabel('Observed Yield')
plt.text(30, 95, "RMSE: 15.19\nR2: 0.711", size=14, va="baseline", ha="right", multialignment="left")
plt.show()
```

## Elastic Net with Different Alpha Value

Use a loop to get RMSE and R2 at different alpha values. 

```python
import warnings
warnings.filterwarnings('ignore') #These two lines are to remove warning information. You can delete this

from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error, r2_score

alpha_list = [0, 0.01, 0.1, 0.2, 0.5, 1]
RMSE_result, R2_result = [], []
for alpha in alpha_list:
    elastic_net = ElasticNet(alpha=alpha, l1_ratio=0.5, random_state=42)
    elastic_net.fit(X_train, y_train)
    y_pred_elastic_net = elastic_net.predict(X_test)
    RMSE_elastic_net = np.sqrt(mean_squared_error(y_test, y_pred_elastic_net))
    R2_elastic_net = r2_score(y_test, y_pred_elastic_net)
    RMSE_result.append(RMSE_elastic_net)
    R2_result.append(R2_elastic_net)
    
print((RMSE_result, R2_result))
```

Now plotting these RMSE and R2 data.

```python
plt.title('Elastic Net')

plt.subplot(121)
plt.plot(RMSE_result, alpha_list , 'b.')
plt.axis([15, 19, -0.1, 1.1])
plt.xlabel('RMSE')
plt.ylabel('alpha')

plt.subplot(122)
plt.plot(R2_result, alpha_list , 'b.')
plt.axis([0.5, 0.75, -0.1, 1.1])
plt.xlabel('R2')
plt.ylabel('alpha')

plt.tight_layout()
plt.show()
```

# Dimension Reduction

For linear regression, too many features can easily lead to overfit. This is why we need to do the dimension reduction. The goal of dimension reduction is to reduce the features we need to train the machine learning model.

This part is a little bit different with the paper, but the goal is same.

First, let's check the most commonly used method PCA. 

## Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
```

For the 'n_components', it can be set in two ways: 1) the number of component (interger), such as 20 or 25 to set how many features we want to use; 2) a float between 0.0 to 1.0, indicating the ratio of variance you want to preserve.

We use the second setting method.

```python
pca = PCA(n_components=0.95, random_state=42)
```

Fit the whole X_scaled data.

```python
X_reduced = pca.fit_transform(X_scaled)
```

Check how many features are left.

```python
pca.n_components_
```

Validate the ratio of variance match the number we set (0.95).

```python
np.sum(pca.explained_variance_ratio_)
```

Check the ratio of each component/feature

```python
pca.explained_variance_ratio_
```

Check the ratio of each component/feature

```python
pca.explained_variance_ratio_
```

Use the reduced X dataset to run a linear regression again, and check the model performance.

```python
from sklearn.model_selection import train_test_split
X_train_reduced, X_test_reduced, y_train_reduced, y_test_reduced = train_test_split(X_reduced, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lin_reg_reduced = LinearRegression()
lin_reg_reduced.fit(X_train_reduced, y_train_reduced)
y_pred_lin_reduced = lin_reg_reduced.predict(X_test_reduced)

RMSE_lin_reduced = np.sqrt(mean_squared_error(y_test_reduced, y_pred_lin_reduced))
R2_lin_reduced = r2_score(y_test_reduced, y_pred_lin_reduced)

print((RMSE_lin_reduced, R2_lin_reduced))
```

```python
plt.plot(y_pred_lin_reduced, y_test_reduced, 'b.')
plt.title('PCA')
plt.axis([-10, 120, -10, 120])
plt.xlabel('Predicted Yield')
plt.ylabel('Observed Yield')
plt.text(30, 95, "RMSE: 20.18\nR2: 0.490", size=14, va="baseline", ha="right", multialignment="left")
plt.show()
```

The PCA dimension reduction didn't imporve the model performance, actually the new model performed worse.

Besides linear regression, PCA can also be used for other non-linear regression methods (SVM, random forest, etc.) with different kernels. It's called kernel PCA (kPCA), and we will use this kPCA later.
